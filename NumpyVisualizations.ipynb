```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
athletes = pd.read_csv("input/athlete_events.csv")
regions = pd.read_csv("input/noc_regions.csv")
```

```python
athletes.head()
```

```python
regions.head()
```

```python
# Join the Dataframes

athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df
```

```python
athletes_df.shape
```

```python
# Column name consistent
athletes_df.rename(columns = {'region' : "Region", 'notes' : 'Notes'}, inplace = True)
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
# check Null values
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# Total number of bull values
athletes_df.isnull().sum()
```

```python
# Q) Print the columns names containing null values or missing values in the form of a list.
athletes_null_culumns = athletes_df.columns[athletes_df.isnull().any()].tolist()
athletes_null_culumns
```

```python
# Nigeria Details using query function 

athletes_df.query('Team == "Nigeria"').head()
```

```python
# Jamaica Details using query function
athletes_df.query('Team == "Jamaica"').head()
```

```python
# Top Countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries
plt.figure(figsize = (12, 6))
plt.xticks(rotation = 20)
plt.title('Overall Participation by Country')
sns.barplot(x = top_10_countries.index, y = top_10_countries, palette = 'Set2');
```

```python
# Age Disctibution of the Participants
plt.figure(figsize = (12, 6))
plt.title("Age Distribution of the athletes")
plt.xlabel('Age')
plt.ylabel("Number of Participants")
plt.hist(athletes_df.Age, bins = np.arange(10, 80, 2), color = 'orange', edgecolor = 'white');
```

```python
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Summer olympics sports
summer_sports = athletes_df[athletes_df.Season == "Summer"].Sport.unique()
summer_sports
```

```python
# Male and Female Participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for male and female athletes (autopct variable enable to display the percent value usign python)
plt.figure(figsize = (12, 6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle=150, shadow = True);
```

```python
# Total medals

athletes_df.Medal.value_counts()
```

```python
# Total number of female athletes in each olympics.

female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
menOlympics = athletes_df[(athletes_df.Sex == 'M') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize = (20, 10))
sns.countplot(x = 'Year', data = womenOlympics, palette = "Spectral")
plt.title('Female Participation')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
part2 = menOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:, 'F'].plot(label='Female')
part2.loc[:, 'M'].plot(label='Male')
plt.title('Male & Female Athletes over time')
plt.legend()
```

```python
# Gold medal Jamaican athletes 
goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals
goldmedals_for_jamaica = athletes_df[(athletes_df.Medal == 'Gold') & (athletes_df.Team == 'Jamaica')]
goldmedals_for_jamaica.head(10)
```

```python
# Take only the values that are different from NaN.

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Gold beyond age of 60 years
goldMedals['ID'][goldMedals['Age'] >= 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] >= 60]
sporting_event
```

```python
# Plot for sporting event
plt.figure(figsize=(10, 5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medals for Athletes over or equal to 60 years of age.")
```

```python
# Gold medals for each country
goldMedals_world = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals_world.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals = goldMedals_world.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index', y = 'Medal', data = totalGoldMedals, height=5, kind="bar", palette= "rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
```

```python
# Silver medals for each country
silverMedals_world = athletes_df[(athletes_df.Medal == 'Silver')]
silverMedals_world.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalSilverMedals = silverMedals_world.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index', y = 'Medal', data = totalSilverMedals, height=5, kind="bar", palette= "rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Silver Medals per Country")

# Bronze medals for each country
bronzeMedals_world = athletes_df[(athletes_df.Medal == 'Bronze')]
bronzeMedals_world.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalBronzeMedals = bronzeMedals_world.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index', y = 'Medal', data = totalBronzeMedals, height=5, kind="bar", palette= "rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Bronze Medals per Country")
```
